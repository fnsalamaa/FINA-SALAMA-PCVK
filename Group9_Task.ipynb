```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image, ImageDraw
import matplotlib.pylab as plt
```

```python
image_4 = cv.imread('ktp.jpeg')
image_5 = cv.cvtColor(image_4, cv.COLOR_BGR2RGB)
cv2_imshow(image_4)
plt.imshow(image_5)
```

```python
img_4 = np.copy(image_4)
#cover Province
for y in range (70,220):
   for x in range (3,10):
     img_4[x,y] = [240,128,128]
#cover Regency
img_4a = np.copy(img_4)
for y in range (95,200):
  for x in range (11,20):
    img_4a[x,y] = [218,112,214]
final_frame = cv.hconcat((img_4,img_4a))
cv2_imshow(final_frame)
```
